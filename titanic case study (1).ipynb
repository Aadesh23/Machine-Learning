## Preprocessing

```python
le=LabelEncoder()
```

```python
co_cols=df.select_dtypes(include='O').columns
co_cols
```

```python
for i in co_cols:
    df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
df=df.drop(["PassengerId", "Name", "Ticket", "Initial","age_group"],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x=df.drop("Survived", axis=1)
y=df["Survived"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=777)
print('x_train shape',x_train.shape,'y_train shape',y_train.shape)
print('y_test shape',x_test.shape,'y_test shape',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB,MultinomialNB
from sklearn.ensemble import AdaBoostClassifier
reg=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
b=BernoulliNB()
m=MultinomialNB()
ad=AdaBoostClassifier()
```

```python
from sklearn.metrics import classification_report
def my_model(model):
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    print('Train Data')
    print(classification_report(y_train,y_pred_train))
    print('Test Data')
    print(classification_report(y_test,y_pred_test))
    return model
```

```python
for i in [reg,knn,dt,rf,b,m,ad]:
    print('when my model is:',i)
    my_model(i)
    print('*'*90)
```

## As you can see Adaboost and Logistic Regression algorithms delivers a perfect accuracy score than the rest of algorithms
